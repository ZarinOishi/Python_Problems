```python
# 6
T = int(input())

for i in range(1,T+1): 
  num = int(input())
  reminder = num % 10
  while num > 9:
    num  = num // 10
  sum = reminder + num
  print("Sum  =", sum)
```

```python
# 8
T = int(input())
for i in range(1, T+1):
  a , b , c = map(int,(input().split()))
  print("case",i,":" )
  temp = 0
  if a > b & a > c :
    temp = a
    a = c
    c =temp      
  
  if a > b :
    temp = a
    a = b
    b =temp
  
  if b > c :
    temp = b
    b = c
    c =temp
  print(a , b , c)

  
```

```python
# 9
import math
T = int(input())
for i in range(1,T+1): 
  num = int(input())
  sq_root = math.sqrt(num)
  if sq_root * sq_root == num :
    print("YES")
  else:
    print("NO")
```
